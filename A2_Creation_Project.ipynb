#  made an object of translator 

```python
translator = Translator()  
```

# First I Translate the header and apply it on whole columns.

```python
df.columns = [translator.translate(col, dest='en').text for col in df.columns]
```

# apply that translator on each  and every text parameter using lambda function 

```python
df = df.applymap(lambda x: translator.translate(str(x), dest='en').text if pd.notnull(x) else x) 
print(df.head())
```
